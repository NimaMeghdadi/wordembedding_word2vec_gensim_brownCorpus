```python
import gensim.models
import logging
from nltk.corpus import brown
from gensim.test.utils import datapath
from gensim import utils
```

```python
model = gensim.models.Word2Vec.load("./tmp/brown_model/word2vec.model")
```

```python
logging.basicConfig(
    format='%(levelname)s : %(message)s', level=logging.INFO)
sentences = brown.sents()
dim = [25, 100, 300]
wind = [2, 7, 15]
neg = [1, 7, 20]

model = gensim.models.Word2Vec(
    sentences,
    # min_count=1,
    vector_size=300,#Dimention
    window=15,
    negative=20,  # negatove>0 -> negative sampling
    # hs=0,  # hs=0
    epochs=5,
    # sg=1,  # sg=1 ->skip-gram
)
model.save("./tmp/brown_model/word2vec.model")
```

```python
model.wv.evaluate_word_pairs('./evaluation/win353.tsv')
```
